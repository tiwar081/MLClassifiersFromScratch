```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import io
import scipy.linalg as la
import pandas as pd
import random as rand
```

This is a visualization of various normalizations for the purpose of regularization during gradient descent.

```python
def isocontours_ln(N, x, y):
    for n in N:
        X, Y = np.meshgrid(x, y)
        Z = np.zeros(X.shape)
        for i in range(len(Z)):
            for j in range(len(Z[0])):
                Z[i][j] = np.linalg.norm(np.array([X[i][j], Y[i][j]]), ord=n)
        contour = plt.contour(X, Y, Z)
        plt.clabel(contour, inline=True, fontsize=8)
        plt.xlabel("x")
        plt.ylabel("y")
        plt.title(f"l-{n} Norm Isocontour Plot")
        plt.gca().set_aspect('equal', adjustable='box')
        plt.show()

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
isocontours_ln([0.5, 1, 2], x, y)
```
